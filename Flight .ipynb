```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('Flight_Train.xlsx')
df1 = pd.read_excel('Flight_set.xlsx')
```

```python
df.head()
```

```python
df1.head()
```

```python
final_df= df.append(df1)
final_df.head()
```

```python
final_df.info()
```

```python
#Feature Engineering
final_df['Date']= final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
final_df['Date']
```

```python
final_df['Month']= final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']= final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Date']= final_df['Date'].astype(int)
final_df['Month']= final_df['Month'].astype(int)
final_df['Year']= final_df['Year'].astype(int)
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df
```

```python
#separarting time and date in Arrival_Time columnn
final_df['Arrival_Time']= final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
#to check no. of null values
final_df.isnull().sum()
```

```python
#separarting time and date in Arrival_Time columnn
final_df['Arrival_Hour']= final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Hour'].head()
```

```python
#separarting time and date in Arrival_Time columnn
final_df['Arrival_Minutes']= final_df['Arrival_Time'].str.split(':').str[1]
final_df['Arrival_Minutes'].head()
```

```python
#converting data type object to integer
final_df['Arrival_Hour']= final_df['Arrival_Hour'].astype(int)
```

```python
#converting data type object to integer
final_df['Arrival_Minutes']= final_df['Arrival_Minutes'].astype(int)
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.info()
```

```python
#separating hor and minute in Dep_Time
final_df['Dep_Hour']= final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minutes']= final_df['Dep_Time'].str.split(':').str[1]
```

```python
#converting both data columns from object to integer
final_df['Dep_Hour']= final_df['Dep_Hour'].astype(int)
final_df['Dep_Minutes']= final_df['Dep_Minutes'].astype(int)
```

```python
final_df.head()
```

```python
#removing Dep_Time column
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
#converting the no. of stops into float 
final_df['Total_Stops']= final_df['Total_Stops'].map({'non-stop':0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan':1})
```

```python
final_df.head()
```

```python
#removing Route column
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration_hour']= final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.info()
```
